```
import matplotlib.pyplot as plt
```

```
x=[2,9,3,7]
y=[7,15,5,9]
plt.xlabel("distance")
plt.ylabel("time")
plt.plot(x,y,label='railway track')
plt.scatter(x,y,s=200,label='railway station',marker='*',color='red')
plt.bar(x,y,label='hight of railway station')
plt.grid(color='green')
plt.legend()
plt.show()
```
